```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy import signal
import pandas as pd
```

```python
#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
t = np.linspace(0,np.pi*100,np.int64(1e5))
```

```python
N = 10+np.sin(t)
```

```python
bkgd = stats.norm.rvs(size=np.int64(1e5))*np.sqrt(N)+N
```

```python
bkgd[0:10]
```

## Problem 1

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t[0:1000],bkgd[0:1000],'o')
plt.show()
```

```python
def plot_range(x_array,y_array, start,stop,step):
    x=x_array[start:stop:step]
    y=y_array[start:stop:step]
    return x,y
```

```python
x,y=plot_range(t,bkgd,0,2000,2)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x,y,'o')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t[::100],bkgd[::100],'o')
plt.show()
```

## Problem 2

```python
fig, ax = plt.subplots(1, 1)
h=ax.hist2d(t,bkgd,bins=(100,100))
plt.colorbar(h[3],ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
h=ax.hist2d(t,bkgd,bins=(300,300))
plt.colorbar(h[3],ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
h=ax.hist2d(t,bkgd,bins=(50,50))
plt.colorbar(h[3],ax=ax)
plt.show()
```

Plotting the 2d histogram displays the density of data points  Starting with the most dense being yellow at the top of the colorbar down to the purple in the colorbar being the least dense. In comparison to the last scatter plot this allows us to see where are the data points are plotted and concentrated rather then just the large area of the same color.
